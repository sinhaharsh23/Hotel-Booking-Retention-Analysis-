## Step 4 — Handle Missing Values



```python
# Check missing values BEFORE
print("=== Missing Values BEFORE Cleaning ===")
miss = df.isnull().sum()
miss_pct = (miss / len(df) * 100).round(2)
report = pd.DataFrame({'Count': miss, 'Pct %': miss_pct})
report = report[report['Count'] > 0].sort_values('Pct %', ascending=False)
print(report)
```

```python
# Fix missing values
df['children'] = df['children'].fillna(0)
df['country']  = df['country'].fillna('Unknown')
df['agent']    = df['agent'].fillna(0)
df['company']  = df['company'].fillna(0)

print("=== After Cleaning ===")
print(f"Total nulls remaining: {df.isnull().sum().sum()}")
print("All missing values handled!")
```

---
## Step 5 — Remove Duplicates & Invalid Rows


```python
# Duplicates
before = len(df)
df = df.drop_duplicates()
print(f"Duplicates removed  : {before - len(df)} rows")

# Invalid: 0 guests
before = len(df)
df = df[~((df['adults']==0) & (df['children']==0) & (df['babies']==0))]
print(f"Zero-guest rows removed: {before - len(df)} rows")

# ADR outliers
before = len(df)
df = df[(df['adr'] >= 0) & (df['adr'] < 5500)]
print(f"ADR outliers removed: {before - len(df)} rows")

print(f"\nFinal dataset size: {len(df):,} rows")
```

---
## Step 6 — Fix Data Types


```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['children'] = df['children'].astype(int)
df['agent']    = df['agent'].astype(int)
df['company']  = df['company'].astype(int)

print("Data types fixed!")
print(df[['reservation_status_date','children','agent','company']].dtypes)
```

---
## Step 7 — Feature Engineering



```python
# 1. Total Guests
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# 2. Total Nights
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# 3. Total Revenue
df['total_revenue'] = (df['adr'] * df['total_nights']).round(2)

# 4. Arrival Date (proper date column)
month_map = {
    'January':1,'February':2,'March':3,'April':4,
    'May':5,'June':6,'July':7,'August':8,
    'September':9,'October':10,'November':11,'December':12
}
df['arrival_month_num'] = df['arrival_date_month'].map(month_map)
df['arrival_date'] = pd.to_datetime({
    'year':  df['arrival_date_year'],
    'month': df['arrival_month_num'],
    'day':   df['arrival_date_day_of_month']
}, errors='coerce')

print("Features 1-4 created: total_guests, total_nights, total_revenue, arrival_date")
```

```python
# 5. Season
def get_season(m):
    if   m in [12,1,2]:  return 'Winter'
    elif m in [3,4,5]:   return 'Spring'
    elif m in [6,7,8]:   return 'Summer'
    else:                return 'Autumn'

df['season'] = df['arrival_month_num'].apply(get_season)

# 6. Traveler Type
def traveler_type(row):
    if   row['company'] != 0:        return 'Corporate'
    elif row['total_guests'] >= 3:   return 'Group/Family'
    elif row['lead_time'] <= 7:      return 'Last-Minute'
    else:                            return 'Leisure'

df['traveler_type'] = df.apply(traveler_type, axis=1)

# 7. Is Weekend Booking?
df['is_weekend_booking'] = (df['stays_in_weekend_nights'] > 0).astype(int)

# 8. Has Special Request?
df['has_special_request'] = (df['total_of_special_requests'] > 0).astype(int)

print("Features 5-8 created: season, traveler_type, is_weekend_booking, has_special_request")
print()
print("=== Preview of All New Features ===")
new_cols = ['total_guests','total_nights','total_revenue','season',
            'traveler_type','is_weekend_booking','has_special_request']
df[new_cols].head(10)
```
